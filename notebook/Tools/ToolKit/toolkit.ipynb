### TOOLKIT

#### A toolkit is just a collection (bundle) of related tools that serve a common purpose packaged together for convenience reusability.

##### 1. A toolkit might be: GoogleDriveToolKit

##### 2. It can contain the following tools

```python
from langchain_core.tools import tool
```

```python
@tool
def add(a:int , b:int) -> int:
"""Add two numbers"""
return a+b 

@tool
def multiply(a:int , b:int) -> int:
"""multiply two numbers"""
return a*b 
```

```python
class MathTool:
   def get_tools(self):
     return[add,multiply]
```

```python
toolkit=MathTool()
tools = toolkit.get_tools()
```

```python
for tool in tools:
print(tool.name, "=>",tool.description)
```
